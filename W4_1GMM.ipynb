```python
# 1
rng=np.random.default_rng(seed=1001) # seed will give the same random sampling everytime with some indentity number here 1001
X=rng.integers(0,2,size=1000) # sample 10 points from 0 and 2 excluding 2
X.mean()
```

```python
# 2
rng=np.random.default_rng(seed=1001)
X=rng.choice([0,1],size=10000,p=[0.2,0.8])   # this is like giving biasness bsed on the chopice of p so 0,1 with some p
X.mean()
```

```python
# 3
rng=np.random.default_rng(seed=1001)
X=rng.choice([0,1],size=1000,p=[0.7,0.3])
X.mean()
```

```python
# 5
plt.bar([0,1],[X[X==0].size,X[X==1].size],width=0.4,color=['blue','red']) # index conditioning
plt.ylabel('Freq of 0 and 1')
plt.xlabel('0 and 1')
plt.title('Bernoulli')
plt.show()
```

```python
# 5
rng=np.random.default_rng(seed=1001)
size_array=[10,100,1000,10000,100000,1000000]
diff=[]
for size in size_array:
  X=rng.choice([0,1],size=size,p=[0.8,0.2])
  mle=X.mean()
  diff.append(abs(mle-0.2))
diff
```

```python
plt.plot(size_array,diff,linestyle='--',marker='o')
plt.xscale('log')  # changes the scale to log scale for better understanding of x axis
```

```python
#7
def Gaussian(x,mu,sigma):
  return (1/(sigma*np.sqrt(2*np.pi)))*np.exp(-0.5*((x-mu)/sigma)**2) # use matcha.io for better latex arkdown uunderstdning
```

```python
Gaussian(5,3,1.5)
```

```python
#10
param=[(5,5),(5,1.5),(1.5,5),(0.5,0.5)]
#X=np.linspace(a,b,50) equal 50 (default) points between a n b interaval
#plt.plot(X,X**2)
x=np.linspace(0,1,100)
for i,(alpha,beta) in enumerate(param):
  y=(x**(alpha-1))*((1-x)**(beta-1))
  plt.subplot(2,2,i+1) # starts from 1 not 0
  plt.plot(x,y,color='skyblue')
  plt.title(f'$\\alpha={alpha},\\beta={beta}$')
plt.tight_layout()

```
