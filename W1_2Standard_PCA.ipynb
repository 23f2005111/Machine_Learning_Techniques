```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([
    [1,-1,3,5,0],
    [2,-3,4,-5,3]
])
```

```python
print(plt.scatter(x[0],x[1]))
```

```python
rng= np.random.default_rng() # random number generator
mu=np.array([0,0])
cov=np.array([[1,0.9],[0.9,1]])

n=1000
x=rng.multivariate_normal(mu,cov,size=n).T
#print(x)
x.shape     # 2 features, 100 dpoints
```

```python
plt.scatter(x[0],x[1],alpha=0.1)
plt.xlim([-10,10])
plt.ylim([-10,10])
plt.axhline(color='k',linestyle='--')
plt.axvline(color='k',linestyle='--')
plt.show()
```

```python
x=x-x.mean(axis=1).reshape(-1,1)
C=x@x.T/n
eigval,eigvec=np.linalg.eigh(C)
# print(eigval)
# print(eigvec)
w_1=eigvec[:,-1]
w_2=eigvec[:,0]

X=np.linspace(-10,10)
y=w_1[1]/w_1[0]*X
plt.scatter(x[0],x[1],alpha=0.1)
plt.plot(X,y,color='r')
plt.xlim([-10,10])
plt.ylim([-10,10])
plt.axhline(color='k',linestyle='--')
plt.axvline(color='k',linestyle='--')
y=w_2[1]/w_2[0]*X
plt.plot(X,y)
```
